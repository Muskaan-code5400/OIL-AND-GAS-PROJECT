# Reservoir Characterization

For this exercise we are going to focus on the application of Machine Learning Models to  Reservoir Characterization.

Machine learning techniques can greatly assist in reservoir characterization, which involves understanding the subsurface properties and behavior of oil and gas reservoirs. Here are some detailed explanations and examples of how machine learning can be applied in reservoir characterization:

**Seismic Data Interpretation: **Seismic data provides valuable information about the subsurface structures and properties. Machine learning algorithms can be applied to seismic data for tasks such as seismic interpretation, attribute analysis, and seismic facies classification. By training on labeled examples, machine learning models can learn to recognize patterns and interpret seismic attributes, aiding in identifying potential hydrocarbon reservoirs.
Example: Convolutional Neural Networks (CNNs) can be employed to automatically extract features from seismic data. By training on labeled seismic sections where reservoirs are known to be present, the CNN can learn to identify similar patterns in unlabeled seismic data, helping in reservoir identification.

**Well Log Analysis:** Well logs contain measurements of various properties, such as gamma-ray, resistivity, porosity, and permeability, which provide insights into the subsurface characteristics. Machine learning algorithms can analyze well log data to predict missing log curves, estimate rock properties, and identify lithology.
Example: Support Vector Machines (SVMs) can be utilized to classify rock types based on well log data. By training on a set of well logs with known rock types, the SVM can learn to classify new wells based on their log responses, aiding in lithology prediction and reservoir characterization.

**Petrophysical Analysis:** Petrophysical analysis involves estimating reservoir properties, such as fluid saturation, porosity, and permeability, using well log measurements. Machine learning techniques can be applied to well log data to build models that predict these properties more accurately.
Example: Random Forest regression models can be trained on a dataset containing well logs and corresponding core data. The model can then predict reservoir properties, such as porosity and permeability, using new well log data as input. This enables a faster and more accurate estimation of reservoir properties.

**Fluid Identification:** Identifying different fluids within the reservoir, such as oil, gas, and water, is crucial for reservoir characterization. Machine learning algorithms can be employed to analyze well log data, seismic attributes, and production data to determine fluid types and distributions.
Example: Unsupervised learning algorithms, such as clustering techniques like K-means or Gaussian Mixture Models, can be applied to well log data to group similar log responses together. This can help in identifying distinct fluid types within the reservoir and understanding their spatial distribution.

**Geological Facies Analysis:** Geological facies analysis involves identifying and characterizing different sedimentary facies within a reservoir. Machine learning algorithms can analyze various data sources, including well logs, core data, and seismic attributes, to classify and model different facies.
Example: Recurrent Neural Networks (RNNs) can be used to analyze well log data sequences and predict facies classes. By training on a dataset of well logs with corresponding facies labels, the RNN can learn the temporal patterns in the logs and classify facies in new wells.

These examples demonstrate how machine learning can be applied to reservoir characterization. By leveraging large volumes of data and learning patterns from labeled examples, machine learning models can enhance the understanding of reservoir properties, improve predictions, and optimize exploration and production strategies.

# Visualize the rock type

```python
import seaborn as sns
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

Before we plot the well data, let's define a color map so the facies are represented by consistent color in all the plots in this tutorial. We also create the abbreviated facies labels, and add those to the facies_vectors dataframe.

```python
# 1=sandstone  2=c_siltstone   3=f_siltstone
# 4=marine_silt_shale 5=mudstone 6=wackestone 7=dolomite
# 8=packstone 9=bafflestone
facies_colors = ['#F4D03F', '#F5B041','#DC7633','#6E2C00',
       '#1B4F72','#2E86C1', '#AED6F1', '#A569BD', '#196F3D']

facies_labels = ['SS', 'CSiS', 'FSiS', 'SiSh', 'MS',
                 'WS', 'D','PS', 'BS']
#facies_color_map is a dictionary that maps facies labels
#to their respective colors
facies_color_map = {}
for ind, label in enumerate(facies_labels):
    facies_color_map[label] = facies_colors[ind]

def label_facies(row, labels):
    return labels[ row['Facies'] -1]

data.loc[:,'FaciesLabels'] = data.apply(lambda row: label_facies(row, facies_labels), axis=1)
```

```python
sns.set()
sns.pairplot(data.drop(['Well Name','Facies','Formation','Depth','NM_M','RELPOS'],axis=1),
             hue='FaciesLabels', palette=facies_color_map,
             hue_order=list(reversed(facies_labels)))
```

Let's take a look at the data from individual wells in a more familiar log plot form. We will create plots for the five well log variables, as well as a log for facies labels.

```python
def make_facies_log_plot(logs, facies_colors):
    #make sure logs are sorted by depth
    logs = logs.sort_values(by='Depth')
    cmap_facies = colors.ListedColormap(
            facies_colors[0:len(facies_colors)], 'indexed')

    ztop=logs.Depth.min(); zbot=logs.Depth.max()

    cluster=np.repeat(np.expand_dims(logs['Facies'].values,1), 100, 1)

    f, ax = plt.subplots(nrows=1, ncols=6, figsize=(8, 12))
    ax[0].plot(logs.GR, logs.Depth, '-g')
    ax[1].plot(logs.ILD_log10, logs.Depth, '-')
    ax[2].plot(logs.DeltaPHI, logs.Depth, '-', color='0.5')
    ax[3].plot(logs.PHIND, logs.Depth, '-', color='r')
    ax[4].plot(logs.PE, logs.Depth, '-', color='black')
    im=ax[5].imshow(cluster, interpolation='none', aspect='auto',
                    cmap=cmap_facies,vmin=1,vmax=9)

    divider = make_axes_locatable(ax[5])
    cax = divider.append_axes("right", size="20%", pad=0.05)
    cbar=plt.colorbar(im, cax=cax)
    cbar.set_label((17*' ').join([' SS ', 'CSiS', 'FSiS',
                                'SiSh', ' MS ', ' WS ', ' D  ',
                                ' PS ', ' BS ']))
    cbar.set_ticks(range(0,1)); cbar.set_ticklabels('')

    for i in range(len(ax)-1):
        ax[i].set_ylim(ztop,zbot)
        ax[i].invert_yaxis()
        ax[i].grid()
        ax[i].locator_params(axis='x', nbins=3)

    ax[0].set_xlabel("GR")
    ax[0].set_xlim(logs.GR.min(),logs.GR.max())
    ax[1].set_xlabel("ILD_log10")
    ax[1].set_xlim(logs.ILD_log10.min(),logs.ILD_log10.max())
    ax[2].set_xlabel("DeltaPHI")
    ax[2].set_xlim(logs.DeltaPHI.min(),logs.DeltaPHI.max())
    ax[3].set_xlabel("PHIND")
    ax[3].set_xlim(logs.PHIND.min(),logs.PHIND.max())
    ax[4].set_xlabel("PE")
    ax[4].set_xlim(logs.PE.min(),logs.PE.max())
    ax[5].set_xlabel('Facies')

    ax[1].set_yticklabels([]); ax[2].set_yticklabels([]); ax[3].set_yticklabels([])
    ax[4].set_yticklabels([]); ax[5].set_yticklabels([])
    ax[5].set_xticklabels([])
    f.suptitle('Well: %s'%logs.iloc[0]['Well Name'], fontsize=14,y=0.94)
```

```python
# Drop any rows with missing values
data = data.dropna()
```

```python
# Encode categorical variables
le = LabelEncoder()
data['Formation'] = le.fit_transform(data['Formation'])
data['Well Name'] = le.fit_transform(data['Well Name'])
data['FaciesLabels'] = le.fit_transform(data['Facies'])
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
# Now, X_train and X_test contain the features, and y_train and y_test contain the labels.
```

## Training the SVM classifier

Now we use the cleaned and conditioned training set to create a facies classifier.  As mentioned above, we will use a type of machine learning model known as a [support vector machine](https://en.wikipedia.org/wiki/Support_vector_machine).  The SVM is a map of the feature vectors as points in a multi dimensional space, mapped so that examples from different facies are divided by a clear gap that is as wide as possible.  

The SVM implementation in [scikit-learn](http://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html#sklearn.svm.SVC) takes a number of important parameters.  First we create a classifier using the default settings.  

```python
from sklearn import svm

SVM = svm.SVC()
```

Now we can train the classifier using the training set we created above.

```python
SVM.fit(X_train,y_train)
```

Now that the model has been trained on our data, we can use it to predict the facies of the feature vectors in the test set.  Because we know the true facies labels of the vectors in the test set, we can use the results to evaluate the accuracy of the classifier.

```python
predicted_labels = SVM.predict(X_test)
```

We need some metrics to evaluate how good our classifier is doing.  A [confusion matrix](http://www.dataschool.io/simple-guide-to-confusion-matrix-terminology/) is a table that can be used to describe the performance of a classification model.  [Scikit-learn](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html) allows us to easily create a confusion matrix by supplying the actual and predicted facies labels.

The confusion matrix is simply a 2D array.  The entries of confusion matrix `C[i][j]` are equal to the number of observations predicted to have facies `j`, but are known to have facies `i`.  

To simplify reading the confusion matrix, a function has been written to display the matrix along with facies labels and various error metrics.  See the file `classification_utilities.py` in this repo for the `display_cm()` function.

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
print("SVM Accuracy:", accuracy_score(y_test, predicted_labels))
```

```python
print(classification_report(y_test, predicted_labels))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Evaluate the models
print("K-Nearest Neighbors (KNN) Accuracy:", accuracy_score(y_test, y_pred_knn))
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_decision_tree))
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_random_forest))
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from mpl_toolkits.axes_grid1 import make_axes_locatable
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import svm
from sklearn.metrics import accuracy_score, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Hyperparameter Tuning for Random Forest
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

random_forest = RandomForestClassifier(random_state=42) #Setting random_state for reproducibility
grid_search = GridSearchCV(random_forest, param_grid, cv=5, scoring='accuracy') #Using 5-fold cross-validation
grid_search.fit(X_train, y_train)

print("Best hyperparameters:", grid_search.best_params_)
print("Best accuracy:", grid_search.best_score_)

best_random_forest = grid_search.best_estimator_
y_pred_random_forest = best_random_forest.predict(X_test)
print("\nClassification Report for Tuned Random Forest:")
print(classification_report(y_test, y_pred_random_forest))


# Hyperparameter Tuning for SVM
param_grid_svm = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto'] # or provide specific values
}

svm_model = svm.SVC(random_state=42) #Setting random_state for reproducibility
grid_search_svm = GridSearchCV(svm_model, param_grid_svm, cv=5, scoring='accuracy')
grid_search_svm.fit(X_train, y_train)

print("\nBest hyperparameters for SVM:", grid_search_svm.best_params_)
print("Best accuracy for SVM:", grid_search_svm.best_score_)

best_svm = grid_search_svm.best_estimator_
y_pred_svm = best_svm.predict(X_test)
print("\nClassification Report for Tuned SVM:")
print(classification_report(y_test, y_pred_svm))
```

```python
models = {
    'SVM': svm.SVC(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
}
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
results = {}
for name, model in models.items():
    # Cross-validation
    scores = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    rmse_scores = np.sqrt(-scores)
    mean_rmse = rmse_scores.mean()

    # Train and evaluate the model
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)

    results[name] = {
        'Cross-Validated RMSE': mean_rmse,
        'Test RMSE': rmse
    }

# Print the results
for name, result in results.items():
    print(f"{name}:")
    print(f"  Cross-Validated RMSE: {result['Cross-Validated RMSE']:.2f}")
    print(f"  Test RMSE: {result['Test RMSE']:.2f}")
```

```python
# Define the models
models = {
    'SVM': svm.SVC(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
}

# Perform cross-validation and evaluate each model
results = {}
for name, model in models.items():
    # Cross-validation (using accuracy as the scoring metric)
    scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy') #using accuracy
    mean_accuracy = scores.mean()

    # Train and evaluate the model on the test set
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)

    results[name] = {
        'Cross-Validated Accuracy': mean_accuracy,
        'Test Accuracy': accuracy
    }

# Print the results
for name, result in results.items():
    print(f"{name}:")
    print(f"  Cross-Validated Accuracy: {result['Cross-Validated Accuracy']:.2f}")
    print(f"  Test Accuracy: {result['Test Accuracy']:.2f}")
```
